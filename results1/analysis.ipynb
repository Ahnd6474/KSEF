# Scoring
Perm_CO2+w1*CO2/N2+w2*CO2/O2+



```python
import numpy as np
import pandas as pd

PATH='mixture_grid_search.csv'

df=pd.read_csv(PATH)
# df: 이미 로드된 DataFrame 이라고 가정
w1,w2=0.78,0.2
import numpy as np
import pandas as pd

def sort_and_save_top10(df: pd.DataFrame, w1: float, w2: float,
                        out_path: str = "top10.csv") -> pd.DataFrame:
    """
    score = Perm_CO2 + w1*(CO2/N2) + w2*(CO2/O2) 로 점수 매기고
    상위 10개를 CSV로 저장한 뒤, 그 DataFrame을 반환.
    """
    eps = 1e-12  # 0 나누기 방지

    co2 = df["perm_CO2"]
    n2 = df["perm_N2"].clip(lower=eps)
    o2 = df["perm_O2"].clip(lower=eps)

    df = df.copy()
    df["score"] = co2 - (n2) -(o2)

    df_sorted = df.sort_values("score", ascending=False).reset_index(drop=True)

    top10 = df_sorted.head(10)
    top10.to_csv(out_path, index=False)

    return top10

# 사용 예시
top10 = sort_and_save_top10(df, w1, w2, out_path=f"top10_candidates.csv")
```
