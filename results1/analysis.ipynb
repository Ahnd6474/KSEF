# Scoring
Perm_CO2+w1*CO2/N2+w2*CO2/O2+



```python
import numpy as np
import pandas as pd

PATH='mixture_grid_search.csv'

df=pd.read_csv(PATH)
# df: 이미 로드된 DataFrame 이라고 가정
w1,w2=0.78,0.2
def sort_by_perm_score(df: pd.DataFrame, w1: float, w2: float) -> pd.DataFrame:
    """
    df를 Perm_CO2 + w1*(CO2/N2) + w2*(CO2/O2) 점수로 정렬해서 반환.
    칼럼 이름:
      - Perm_CO2 : CO2 투과도
      - Perm_N2  : N2 투과도
      - Perm_O2  : O2 투과도
    """
    eps = 1e-12  # 0으로 나누기 방지용

    co2 = df["perm_CO2"]
    n2 = df["perm_N2"].clip(lower=eps)
    o2 = df["perm_O2"].clip(lower=eps)

    df = df.copy()
    df["score"] = 0.5* co2 - w1 * (n2) - w2 * (o2)

    # 점수 높은 순으로 정렬
    df_sorted = df.sort_values("score", ascending=False).reset_index(drop=True)
    return df_sorted
# 점수 컬럼 추가
df_sorted = sort_by_perm_score(df, w1=0.5, w2=1.0)
# 점수 기준으로 정렬 (내림차순: 점수가 큰 순서)
df_sorted.head()
```
